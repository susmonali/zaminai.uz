```python
import ee
import geemap

ee.Initialize(project="affable-tribute-476418-a4")

aoi = ee.Geometry.Rectangle([71.610947,40.362242,71.848526,40.490174]) 

collection = (ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
    .filterBounds(aoi)
    .filterDate("2026-01-01", "2026-01-31")
    .filter(ee.Filter.lt("CLOUDY_PIXEL_PERCENTAGE", 10))
)

print("Images found:", collection.size().getInfo())

image = collection.median().clip(aoi)
```

```python
import geemap
image = collection.median().clip(aoi)

Map = geemap.Map(center=[71.738148,40.411405], zoom=8)

vis_true = {"bands": ["B4", "B3", "B2"], "min": 0, "max": 3000}
Map.addLayer(image, vis_true, "True color")

vis_false = {"bands": ["B8", "B4", "B3"], "min": 0, "max": 4000}
Map.addLayer(image, vis_false, "False color (vegetation)")

Map.addLayerControl()
Map
```
